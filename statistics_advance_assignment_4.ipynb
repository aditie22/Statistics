Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

Both the t-test and the z-test are statistical tests used for hypothesis testing, particularly for comparing means of two groups. However, they are used in different situations based on the information available and the assumptions about the population parameters. Here's the difference between the two along with example scenarios for each:

1. T-test:
The t-test is used when the population standard deviation is unknown and must be estimated from the sample. It is more suitable for small sample sizes (typically n < 30) and when the underlying population distribution is approximately normal. There are different types of t-tests, including the independent samples t-test and the paired samples t-test.

Example Scenario:
Suppose you are testing a new drug's effectiveness in reducing blood pressure. You have a sample of 20 patients who were given the new drug, and you want to compare their average reduction in blood pressure to a control group of 20 patients who were given a placebo. Since you don't know the population standard deviation of blood pressure reduction, you'd use an independent samples t-test to determine if there's a significant difference between the two groups' mean reductions.

2. Z-test:
The z-test is used when the population standard deviation is known or when the sample size is sufficiently large (typically n ≥ 30), allowing for the use of the standard normal distribution for calculations. It is applicable when the population distribution is approximately normal or when the sample size is large enough to invoke the Central Limit Theorem.

Example Scenario:
Imagine you work at a company that produces light bulbs. You want to test if a new manufacturing process increases the average lifespan of light bulbs. You take a sample of 100 light bulbs produced using the new process and another sample of 100 bulbs produced using the old process. If you know the population standard deviation of the bulb lifespans (which is rare in practice), you could use a z-test to determine if the new process leads to a statistically significant increase in the average lifespan of the light bulbs.

In summary, use a t-test when the population standard deviation is unknown and the sample size is small, and use a z-test when the population standard deviation is known or the sample size is large. It's important to choose the appropriate test based on the specific characteristics of your data and the assumptions you can make about the underlying population distribution.

Q2: Differentiate between one-tailed and two-tailed tests.

In statistical hypothesis testing, both one-tailed tests and two-tailed tests are used to assess the validity of a hypothesis about a population parameter based on sample data. They differ in their approach to evaluating whether the sample evidence supports or contradicts the null hypothesis.

One-Tailed Test:
A one-tailed test, also known as a one-sided test, is used when the research hypothesis specifies a directional relationship between the variables being tested. In other words, it is used when you're interested in detecting an effect in only one direction. The critical region for the test is located entirely in one tail of the probability distribution curve.
For example, if you're testing whether a new drug increases reaction times, your null hypothesis (H0) might be that the drug has no effect, and your alternative hypothesis (Ha) might be that the drug increases reaction times. In this case, you would perform a one-tailed test to see if the data supports the idea that reaction times are increased.

The critical region is defined based on the chosen significance level (alpha) and the direction specified by the alternative hypothesis. If the test statistic falls into this critical region, you would reject the null hypothesis in favor of the alternative.

Two-Tailed Test:
A two-tailed test, also known as a two-sided test, is used when the research hypothesis does not specify a particular direction of the effect, or when you are interested in detecting effects in both directions. The critical region for the test is divided between both tails of the probability distribution curve.
For instance, if you're testing whether a new treatment affects participants' weight, your null hypothesis (H0) might be that the treatment has no effect on weight, and your alternative hypothesis (Ha) might be that the treatment does have an effect on weight, regardless of whether it causes weight gain or weight loss. In this case, you would perform a two-tailed test to account for the possibility of effects in both directions.

The critical regions are again determined by the chosen significance level (alpha), but they are divided into two parts – one in each tail of the distribution. If the test statistic falls into either of these critical regions, you would reject the null hypothesis.

In summary, the main difference between one-tailed and two-tailed tests lies in the directional nature of the hypotheses being tested. One-tailed tests focus on effects in a single direction specified by the alternative hypothesis, while two-tailed tests are used when effects in either direction are of interest. The choice between these tests depends on the specific research question and the hypotheses being investigated.

Q3.Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

In hypothesis testing, Type 1 and Type 2 errors are two possible mistakes that researchers can make when interpreting the results of a statistical test. These errors are associated with the decisions made based on the null hypothesis (H0) and the alternative hypothesis (H1).

Type 1 Error (False Positive): A Type 1 error occurs when a researcher rejects the null hypothesis (H0) when it is actually true. In other words, the researcher concludes that there is a significant effect or difference when, in reality, there is no such effect or difference in the population being studied. This error is also referred to as a "false positive."

Example Scenario: Imagine a clinical trial for a new drug where the null hypothesis is that the drug has no effect on patients' recovery time. If the researchers incorrectly reject the null hypothesis and conclude that the drug is effective, but in reality, the drug has no effect, it would be a Type 1 error.

Type 2 Error (False Negative): A Type 2 error occurs when a researcher fails to reject the null hypothesis (H0) when it is actually false. In other words, the researcher fails to detect a significant effect or difference when there actually is such an effect or difference in the population being studied. This error is also referred to as a "false negative."

Example Scenario: Continuing with the drug trial example, let's say the null hypothesis is that the drug is effective at reducing recovery time, but in reality, the drug does have a positive effect. If the researchers fail to reject the null hypothesis and conclude that the drug is not effective, even though it actually is, it would be a Type 2 error.

In hypothesis testing, researchers aim to strike a balance between these two types of errors. The level of significance, often denoted by α (alpha), determines the threshold for making a decision about the null hypothesis. Lowering the level of significance reduces the chance of making a Type 1 error but increases the chance of making a Type 2 error, and vice versa.

It's important to note that the probabilities of Type 1 and Type 2 errors are interconnected; as one decreases, the other increases. This trade-off is inherent in hypothesis testing and requires careful consideration depending on the consequences of making each type of error in a specific context.
